```python
columnas = ["SepalLengthCm","SepalWidthCm","PetalLengthCm", "PetalWidthCm","Species"]
df = df[columnas]
```

```python
df
```

```python
# Crear toda la lista de variables categóricas para transformarlas a numéricas
todummy_list = ["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]
```

```python
# Función para llamar get dummies c
def dummy_df(df, todummy_list):
    for x in todummy_list:
        dummies = pd.get_dummies(df[x], prefix=x, dummy_na=False)
        df = df.drop(x, 1)
        df = pd.concat([df, dummies], axis=1)
    return df
```

```python
df = dummy_df(df, todummy_list)
df
```

```python
print(df)
```

```python
# Crear dos DataFrames X con variables indpendientes o features y uno y con variable dependiente , la que vamos a predecir
X = df.drop('Species',1)
Y = df.Species
```

```python
from sklearn.feature_selection import SelectKBest, chi2

select = SelectKBest(chi2,k=55) #124 es el total, 62 es la mitad y 42 es la tercera parte
selected_features = select.fit(X, Y)
indices_selected = selected_features.get_support(indices=True)
colnames_selected = [df.columns[i] for i in indices_selected]

X = df[colnames_selected]
X
```

```python
from sklearn.model_selection import train_test_split
#Separación los datos de "train" en entrenamiento y prueba para probar los algoritmos
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
print('Son {} datos para entrenamiento y {} datos para prueba'.format(X_train.shape[0], X_test.shape[0]))
```
